```python
#Define the inspect_dataset function, then call
def inspect_dataset(df, name):
    print(f"\n--- {name} Dataset ---")
    print(f"Shape: {df.shape}")
    print("\nFirst few rows:")
    print(df.head())
    print("\nData types:")
    print(df.dtypes)
    print("\nMissing values:")
    print(df.isnull().sum())
    print("\nBasic statistics:")
    print(df.describe())
```

```python
# Inspect each dataset
inspect_dataset(customers, "Customers")
inspect_dataset(order_details, "Order Details")
inspect_dataset(orders, "Orders")
inspect_dataset(products, "Products")
```

```python
# Convert date columns to datetime
orders['orderDate'] = pd.to_datetime(orders['orderDate'])
orders['requiredDate'] = pd.to_datetime(orders['requiredDate'])
orders['shippedDate'] = pd.to_datetime(orders['shippedDate'])

print(orders[['orderDate', 'requiredDate', 'shippedDate']].dtypes)
print(orders[['orderDate', 'requiredDate', 'shippedDate']].head())
```

```python
# missing values
print("Missing values before:", orders['shippedDate'].isnull().sum())

# Fill missing shippedDate with requiredDate (you can adjust this strategy if needed)
orders['shippedDate'] = orders['shippedDate'].fillna(orders['requiredDate'])

# Check missing values after filling
print("Missing values after:", orders['shippedDate'].isnull().sum())
```

```python
def check_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    print(f"\nOutliers in {column}:")
    print(outliers[column])
    return outliers

# Check outliers in freight
freight_outliers = check_outliers(orders, 'freight')

# Check outliers in unitPrice
unitprice_outliers = check_outliers(products, 'unitPrice')
```

```python
# Check data types of ID columns
print("\nData types of ID columns:")
print("customers['customerID']:", customers['customerID'].dtype)
print("orders['customerID']:", orders['customerID'].dtype)
print("orders['orderID']:", orders['orderID'].dtype)
print("order_details['orderID']:", order_details['orderID'].dtype)
print("products['productID']:", products['productID'].dtype)
print("order_details['productID']:", order_details['productID'].dtype)

# Convert if necessary (in this case, it seems unnecessary, but included for completeness)
orders['customerID'] = orders['customerID'].astype(str)
customers['customerID'] = customers['customerID'].astype(str)
```

```python
# Merge datasets
merged_data = orders.merge(order_details, on='orderID')
merged_data = merged_data.merge(customers, on='customerID')
merged_data = merged_data.merge(products, on='productID')

# Check the shape and columns of the merged dataset
print("\nShape of merged dataset:", merged_data.shape)
print("\nColumns in merged dataset:", merged_data.columns)

# Display the first few rows of the merged dataset
print("\nFirst few rows of merged dataset:")
print(merged_data.head())
```

```python
# Rename unitPrice columns
merged_data = merged_data.rename(columns={'unitPrice_x': 'order_unit_price', 'unitPrice_y': 'product_list_price'})

# Add total_order_value field
merged_data['total_order_value'] = merged_data['order_unit_price'] * merged_data['quantity'] * (1 - merged_data['discount'])

# Add days_to_ship field
merged_data['days_to_ship'] = (merged_data['shippedDate'] - merged_data['orderDate']).dt.days

# Display the updated columns and first few rows
print("\nUpdated columns:")
print(merged_data.columns)
print("\nFirst few rows with new fields:")
print(merged_data[['orderID', 'order_unit_price', 'product_list_price', 'quantity', 'discount', 'total_order_value', 'days_to_ship']].head())

# Check the date range of the orders
print("\nDate range of orders:")
print("Earliest order date:", merged_data['orderDate'].min())
print("Latest order date:", merged_data['orderDate'].max())
```

```python
# 1. Customer Loyalty Analysis
customer_order_counts = merged_data.groupby('customerID')['orderID'].nunique().sort_values(ascending=False)
print("Top 10 customers by number of orders:")
print(customer_order_counts.head(10))

# 2. Product Popularity Analysis
product_sales = merged_data.groupby('productName')['quantity'].sum().sort_values(ascending=False)
print("\nTop 10 products by total quantity sold:")
print(product_sales.head(10))

# 3. Customer Segmentation by Total Spend
customer_total_spend = merged_data.groupby('customerID')['total_order_value'].sum().sort_values(ascending=False)
customer_segments = pd.qcut(customer_total_spend, q=3, labels=['Low', 'Medium', 'High'])
customer_segment_df = pd.DataFrame({'customer_segment': customer_segments})

# Merge customer segments back to the main dataframe
merged_data = merged_data.merge(customer_segment_df, left_on='customerID', right_index=True)

# Analyze product preferences by customer segment
segment_product_preferences = merged_data.groupby(['customer_segment', 'productName'])['quantity'].sum().sort_values(ascending=False)
print("\nTop 5 products for each customer segment:")
for segment in ['Low', 'Medium', 'High']:
    print(f"\n{segment} Spend Segment:")
    print(segment_product_preferences[segment].head())

# 4. Average days to ship by product
avg_days_to_ship = merged_data.groupby('productName')['days_to_ship'].mean().sort_values(ascending=False)
print("\nTop 10 products with longest average ship time:")
print(avg_days_to_ship.head(10))

# 5. Analyze order frequency
merged_data['order_month'] = merged_data['orderDate'].dt.to_period('M')
monthly_orders = merged_data.groupby('order_month')['orderID'].nunique()
print("\nMonthly order trend:")
print(monthly_orders)

# 6. Analyze average order value by customer segment
avg_order_value = merged_data.groupby('customer_segment')['total_order_value'].mean()
print("\nAverage order value by customer segment:")
print(avg_order_value)
```
